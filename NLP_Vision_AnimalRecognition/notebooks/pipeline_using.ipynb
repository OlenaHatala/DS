# Verification Pipeline: Matching Text and Images
## Overview
This section demonstrates how to use the VerificationPipeline to compare text-based Named Entity Recognition (NER) results with image classification predictions.
The pipeline will:

Extract animal entities from the text using the NER model.
Classify an animal from an image using the image recognition model.
Verify if the detected text entity matches the predicted image label.
